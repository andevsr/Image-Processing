```python
from sklearn.cluster import KMeans
from collections import Counter
import numpy as np
import cv2                  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main_colors = np.array([   [0,0,0], #Black
                  [255,255,255],      #White
                  [255,0,0],          #Red  
                  [0,255,0],          #Lime
                  [0,0,255],          #Blue
                  [255,255,0],        #Yellow
                  [0,255,255],        #Cyan
                  [255,0,255],        #Magenta
              ] )
```

```python
colours = ["Black","White","Red","Lime","Blue","Yellow","Cyan","Magenta"]
```

```python
def RMSD(a,b):
    return (np.square(a[0]-b[0])+np.square(a[1]-b[1])+np.square(a[2]-b[2]))
```

```python
def colour_detection(img):
    img = cv2.resize(img, (600, 400), interpolation = cv2.INTER_AREA)
    img = img.reshape((img.shape[0]*img.shape[1],3))
    
    df = [0,0,0,0,0,0,0,0]
    for x in img:
        dist = []
        for y in main_colors:
            dist.append(RMSD(x,y))
        dist = [x/8 for x in dist]
        dist = [np.sqrt(x) for x in dist]
        ind = dist.index(min(dist))
        df[ind] += 1
    count = 0
    for x in df:
        if(x != 0):
            count += 1
    return df,count
```

```python
def colour_names(clrs):
    names=[]
    for x in range(8):
        if(clrs[x] != 0):
            names.append(colours[x])
    return names
    
```

```python
image = cv2.imread('Invoice 01.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.axis("off")
plt.imshow(image)
names, num = colour_detection(image)
print("The number of Colours in this Image is {}".format(num))
print("The Colours are: ")
names = colour_names(names)
for x in names:
    print(x)
```
