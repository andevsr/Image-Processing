```python
from scipy import ndimage
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rotate_image(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    can_img = cv2.Canny(img_gray, 100, 200)
    lines = cv2.HoughLinesP(can_img, 1, np.pi / 180.0, 120, minLineLength=100, maxLineGap=10)
    
    for x1, y1, x2, y2 in lines[2]:
        cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        angle = math.degrees(math.atan2(y2 - y1, x2 - x1))
    
    median_angle = np.median(angle)
    print ("Angle is {}".format(median_angle))
    
    img_rotated = ndimage.rotate(image, median_angle)
    
    return img_rotated
```

```python
img_before = cv2.imread('Invoice 01.jpeg')
img_rotated = rotate_image(img_before)
img_rotated = ndimage.rotate(img_rotated, 90)
plt.figure(figsize=(20,10))
plt.imshow(img_rotated)
```
